```python
from queue  import Queue#it has all the locks.Queue is used for resource sharing
```

```python
def producer(q):#method
    for i in range(50):
        q.put(i)#it will put all in q --> queue.
        print("Published", i)
def consumer(q):
    while True:#consume indefinitely to get product
        data = q.get()#get's the last value
        print("consumed",data)
        
```

```python
q = Queue()
```

```python
t_producer = Thread(target= producer, args = (q,))
t_consumer = Thread(target= consumer, args = (q,))
```

```python
t_consumer.start()
```

```python
t_producer.start()
```

```python
t_producer.join()
```

```python
t_consumer.join() #still running because waiting continously
```

#### Couroutines 
- Knowingly gives context to another function.
-  uses yield ---> return the data and consumes the data 

```python
def print_fancy_name(prefix):
    try:
        while True:#this coroutine function is running infinitely
            name1 = (yield)
           #this helps the fuction to consume a value from another
                      #fuction call.
            print(prefix + ":" + name1 )#these name = yeild is coroutine
    except GeneratorExit:
        print("Done!")
    
  #anything before the yield statement is initialization 
    #this funtion return us generators.
```

```python
co= print_fancy_name("cool")
co
```

### Async-IO

```python
import asyncio
import time
```

```python
 def main():
    print("Hello world")
```

```python
if __name__  == '__main__':
    main()
```

```python
async  def main():
    print("Hello world")
    await asyncio.sleep(1)#by 1 sec--->blocking the code , completes the execution 
    #after 1 sec# it wont be giving any coroutine
    print("world")
```

```python
if __name__  == '__main__':
    print(type(main()))
    asyncio.run(main())#as main function only return coroutine , 
#asyncio.run will execute the coroutine . This call is also blocking
#Create a event loop ie global event loop 
    print("program ended")
```
