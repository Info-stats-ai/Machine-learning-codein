### Multivariate Normal Distribution

```python
u = 5# mean
sigma = 1
values = u + sigma*np.random.randn(1000)
plt.hist(values,bins= 100, color= "red")
plt.show()
```

```python
x = values 
y = np.zeros(x.shape)
y
```

```python
plt.scatter(x,y) # u = 5, sigma =1 
plt.show() 3 density around 5 is maximum
```

```python
# spread = sigma-x and sigma-y, mean = ux, uy
```

### Univariate Analysis

```python
mean = np.array([0.0,0.0])
cov  = np.array([[1, 0],[0, 1]]) #here x and y  are independent of each other
dist = np.random.multivariate_normal(mean, cov, 500)
print(dist.shape)# x is in 0th col , y is in 1st col
plt.scatter(dist[:,0],dist[:,1])
```

```python
mean = np.array([0.0,0.0])
cov  = np.array([[1, 0.8],[0.8, 1]]) #here x and y  are dependent of each other
dist = np.random.multivariate_normal(mean, cov, 500)
print(dist.shape)# x is in 0th col , y is in 1st col
plt.scatter(dist[:,0],dist[:,1]) # almost foll wer linear distribution
plt.show()
```

### Bivariate Analysis

```python
mean = np.array([0.0,0.0])
cov  = np.array([[1, 0.8],[0.8, 1]]) #here x and y  are dependent of each other
dist = np.random.multivariate_normal(mean, cov, 500)
print(dist.shape)# x is in 0th col , y is in 1st col
plt.scatter(dist[:,0],dist[:,1]) # almost foll wer linear distribution
mean2 = np.array([5.0,6.0])
cov2  = np.array([[1.3, 0.2],[0.2, 1.1]]) #here x and y  are dependent of each other
dist2 = np.random.multivariate_normal(mean2, cov2, 500)
print(dist.shape, dist2.shape)# x is in 0th col , y is in 1st col
plt.scatter(dist[:,0],dist[:,1])
plt.scatter(dist2[:,0],dist2[:,1])
plt.show()
```

```python
# orange one is less spread than green one(mean is 5, or 6 around)
```
