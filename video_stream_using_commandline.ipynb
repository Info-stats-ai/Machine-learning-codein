### Simple program to read and show an image

```python
# We are going to Read a video Stream from Camera(Frame by Frame) 
# each video is  stream of images
#each image is stream of frame.
import cv2
capture = cv2.VideoCapture(0) # inside the fuction we define id of our video
# if it is zero it will take the default value of it.

while True :
    return_value, frame = capture.read()
#it return two values, one is boolean value and other is frames
    if return_value == False :
        #image is not captured properly
        continue
    cv2.imshow("Video Frame", frame)

#we will wait for user input - user wants to press q,then you will stop the loop


key_pressed_user =cv2.waitKey(1) & 0xFF 

# wait key gives us a 32 bit integer
# and 0xFF has---->>> It has 8 ones 1111111---what we did here we converted 32 bit
# into 8 bit integer by using amparsand, which returns last 8 bits of the 32 bit integer
# and now crd can find the value of 8 bit integer

if key_pressed_user == ord('q'): #checks the Ascii value, gives 8 bit integer
    break

cap.release()
cv2.destroyAllWindows()

    
```
