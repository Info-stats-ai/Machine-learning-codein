### Linear regression Using Scikit learn 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
X, y = make_regression() 
# creating an object of class make regression
print(X.shape, y.shape)
```

```python
X,y= make_regression(n_samples=1000,n_features=2,n_informative=1,noise=10,random_state=1)
print(X.shape,y.shape)

#informative = 1 , means only  1 feature is important though

```

```python
plt.subplot(1, 2, 1)
plt.grid()
plt.scatter(X[:,0], y,) # first feature
plt.subplot(1, 2, 2)
plt.scatter(X[:, 1], y, color ='orange') # second feature
plt.grid()
plt.show()
# from this we can state that  second feature is not a very useful feature.
```

### Implementing LR Using Sk -Learn

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression() # model object 
```

```python
model.fit(X,y) # training my data
# model object contains my parameters after the training, i, e we
# have theta nod and theta one
```

```python
model.coef_ # THETA 1 AND THETA  2 respectively
```

```python
model.intercept_ # theta zero
```

```python
model.predict([X[0]]) # predicted f
```

```python
y[0] # real valuye
```
