#### BOX Plot

```python
sns.boxplot? # bar plot , is for continous  data  and less  stats can be found
```

```python
sns.boxplot(x = 'day', y = 'total_bill', hue = 'sex',data = tips_dataset)
```

##### Box and Viscous 

- 25%, 50% , 75% quartiles
- mid line is the median
- line end are max and minimum respectively
- IQR = difference between 25 and 75th quartiels
- large big  diamond shaped dots are outliers
-  Quartile-3(75%)+ 1.5*IQR = Maximum value
-  Quartile-1(25%)- 1.5*IQR = Minimum value
